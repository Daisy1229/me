```python
if os.path.isfile("mortality_rate.csv"):
    filepath = "mortality_rate.csv"
    print("loading from file")
else:
    filepath = "https://ncci.canceraustralia.gov.au/outcomes/cancer-mortality/cancer-mortality#"
    print("loading from the internet")

mortality_rate= pd.read_csv(filepath)
print("done")
```

```python
mortality_rate.head()
```

```python
f, ax1= plt.subplots(figsize=(10,10))
mortality_rate["Males"].plot() and mortality_rate["Females"].plot() and mortality_rate["Persons"].plot()
ax1.set_title('Age standardised mortality rate, by sex, 1968 to 2015')
ax1.set_xlabel('Year')
ax1.set_ylabel('No.per 100,000')
plt.show()
```

```python
mortality_rate["Males"].max(), mortality_rate["Males"].min()
```

```python
mortality_rate["Females"].max(), mortality_rate["Females"].min() 
```

```python
f, ax1= plt.subplots(figsize=(13,7))
mortality_rate["Females"].hist() and mortality_rate["Males"].hist() 
ax1.set_title('Age standardised mortality rate, by sex, 1968 to 2015')
ax1.set_xlabel('No.per 100,000')
ax1.set_ylabel('Year')
plt.show()
```
